```python
x = df.drop(columns = ['price'],axis = 1)
y = df['price']
x.head()
```

```python
x_one_encoder=pd.get_dummies(x,columns=['street','city','statezip','country','date'],drop_first = True)
x_one_encoder
```

```python
x_one_encoder=x_one_encoder.astype(int)
x_one_encoder
```

```python
from sklearn.preprocessing import StandardScaler
numerical_cols = ['bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'floors','waterfront','view','condition','sqft_above','sqft_basement','yr_renovated','yr_built']
scaler = StandardScaler()
x_one_encoder[numerical_cols] = scaler.fit_transform(x_one_encoder[numerical_cols])

x_one_encoder
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

X_train, X_test, y_train, y_test = train_test_split(x_one_encoder, y, test_size=0.33, random_state=42)

model = LinearRegression()
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)

y_pred
```

```python
y_test
```

```python
r2 = r2_score(y_test,y_pred)
r2
```

```python
n = X_test.shape[0]
p = X_test.shape[1]
adjusted_r2 = 1 - ((1 - r2) * (n - 1)) / (n - p - 1)
print("Adjusted R² Score:", adjusted_r2)
```
